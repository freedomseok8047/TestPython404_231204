```python
# matplotlib 공식 문서 예제 가져오기. 
import matplotlib.pyplot as plt
import numpy as np


species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
# np 슬라이싱 
import numpy as np
import pandas as pd 
import matplotlib

# 3 x 2 행렬,
        #0열 #1열
list = [[1,2], #0행
        [3,4], #1행
        [5,6]] #2행
# 리스트 -> np 로 변환하고, 슬라이싱하기. 
np_array = np.array(list)
#슬라이싱 해보기. [시작값:끝값:증가값], 끝값: 미포함
# 1행, 1열
print(f"np_array[1][1] : \n {np_array[1][1]} ")
# 0행 1행 미포함
print(f"np_array[0:1] : \n {np_array[0:1]} ")
# 0행 1행 포함
print(f"np_array[0:2] : \n {np_array[0:2]} ")
# 전체 행, 1열 
print(f"np_array[:,1] : \n {np_array[:,1]} ")
# 0행 2행 의 1열 부분, 2, 6
print(f"np_array[[0,2],1] : \n {np_array[[0,2],1]} ")
```

```python
# 브로드캐스트, 정렬. 

import numpy as np
import pandas as pd 
import matplotlib

# 3 x 2 행렬,
        #0열 #1열
list = [[1,2], #0행
        [3,4], #1행
        [5,6]] #2행
# 리스트 -> np 로 변환하고, 슬라이싱하기. 
np_array = np.array(list)
# 각 넘파이 배열에 모든 요소에 특정 값 한번에 모두 더하기. : 브로드캐스트 
print(f"np_array 연산전: \n {np_array} ")
print(f"np_array + 100 연산후: \n {np_array + 100} ")

# 각 행, 열의 합, axis=0 , 열기준, axis=1, 행기준
# 헷갈려서, 저는 : " 넘파이는 1행이다. " -> 1 행 기준이라고, 
print(f"np_array 0열 기준의 합: \n {np.sum(np_array, axis=0)} ")

print(f"np_array 0행 기준의 합: \n {np.sum(np_array, axis=1)} ")

# 각 행, 열의 기준 정렬 
print(f"np_array 행 기준 정렬: \n {np.sort(np_array, axis=1)} ")
print(f"np_array 열 기준 정렬: \n {np.sort(np_array, axis=0)} ")
print(f"np_array 열 기준  내림차순 정렬: \n {np.sort(np_array, axis=0)[::-1]} ")
```

```python
# 조건식 

import numpy as np
import pandas as pd 
import matplotlib

# 0~ 255(미포함) 사이에서 램덤한 숫자 선택 후, 조건식 하기. 
arr1 = np.random.randint(0,255,size=(3,3))
print(f"arr1 : \n {arr1} ")

arr2 = np.random.randint(0,255,size=(3,3))
print(f"arr2 : \n {arr2} ")

# 랜덤하게, 특정의 값을 선택하는 형식
condArr = np.random.choice([True,False],size=(3,3))
print(f"condArr : \n {condArr} ")

# 조건식 걸기. where 함수(조건식, 참일경우 수행, 거짓일 경우 수행)
np.where(condArr,arr1,arr2)
```

```python
# 판다스 Series에서 , 리스트, 컬럼스 옵션, zip 함수 이용 
# zip 함수는 같은 인덱스끼리 묶기.

import numpy as np
import pandas as pd 
import matplotlib

data = [["준석1","준석2","준석3"],
        ["메뉴1","메뉴2","메뉴3"],
        ["1000원","2000원","3000원"]]
# zip 함수로 같은 인덱스 묶기
data2 = zip(data[0], data[1] , data[2])
df = pd.DataFrame(data2, columns=["이름","메뉴", "가격"])
print(f"df : \n {df} ")
```

```python

#판다스 데이터 프레임 각 열, 행 선택 및 값 수정.

import numpy as np
import pandas as pd 
import matplotlib

data= [["상용","상용2","상용3"],
       ["메뉴1","메뉴2","메뉴3"],
       ["1000원","2000원","3000원"]]


# zip 함수로 같은 인덱스 묶기 
data2 = zip(data[0],data[1],data[2])

df = pd.DataFrame(data2, columns=["이름","메뉴","가격"])

print(f"df : \n {df} ")
# 각 열 선택 
print(f"df  이름 열 선택 : \n {df["이름"]} ")
# 각 열 값 대입
df["이름"]='lsy'
print(f"df  이름 열 값 수정 후  조회 : \n {df} ")

# 열 삭제. 
del df["이름"]
print(f"df  이름 열 값 삭제 후 조회 : \n {df} ")
```

```python
# 데이터 프레임에서 , 각행의 NaN(Not a Number ) 없애기
# 몇개 있을 때 없애기, 또는 , 전부 NaN 일 경우 없애기. 
import numpy as np
import pandas as pd 
import matplotlib

data= [["상용","상용2","상용3"],
       ["메뉴1","메뉴2","메뉴3"],
       ["1000원","2000원","3000원"],
       [1,2,np.nan],
       [np.nan,np.nan,np.nan]]




df = pd.DataFrame(data, columns=["이름","메뉴","가격"])

print(f"df : \n {df} ")

# np.nan NaN 없애기. 각 행중에서, 하나라도 NaN 있다면 없애기
# df.dropna(how='any') , 수행후 , 처리한 새로운  df 받아서 출력해야하고
# 만약, 기존 데이터 프레임에 바로 적용 할려면, 속성 inplace = True 
#df.dropna(how='any',inplace=True)
df2 = df.dropna(how='any')
print(f"\n df NaN 없애고 새로운 df2 에 할당: \n {df2} ")

df.dropna(how='any',inplace=True)
print(f"\n df NaN 없애고 기존 df 에 표기 : \n {df2} ")
```

```python
# 데이터 프레임에서 , 각행의 NaN(Not a Number ) 없애기
# 몇개 있을 때 없애기, 또는 , 전부 NaN 일 경우 없애기. 
import numpy as np
import pandas as pd 
import matplotlib

data= [["상용","상용2","상용3"],
       ["메뉴1","메뉴2","메뉴3"],
       ["1000원","2000원","3000원"],
       [1,2,np.nan],
       [np.nan,np.nan,np.nan]]




df = pd.DataFrame(data, columns=["이름","메뉴","가격"])

print(f"df : \n {df} ")


# 가격 열에서, NaN 있는 행 추출하기
df3 = df.loc[df.isnull()['가격'],:]
print(f"\n df3 가격 열에서, NaN 있는 행 추출하기: \n {df3} ")

# 특정 행 삭제 , 삭제 후 데이터 담거나 또는 기존 데이터를 사용시 
# inplace=True 사용, axis=0 , 행기준, axis=1 열기준
df.drop(0, axis=0, inplace=True)
print(f"df 0행 삭제 후 조회 : \n {df} ")

```
